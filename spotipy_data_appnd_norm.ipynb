```python
import warnings                                 # Used to ignore the warning given as output of the code
warnings.filterwarnings('ignore')
from collections import defaultdict             # A dictionary output that does not raise a key error

import numpy as np                              # Basic libraries of python for numeric and dataframe computations
import pandas as pd
import os
```

```python
def concat_csv_files(folder_path):
    # Get a list of all CSV files in the folder
    csv_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]
    final_df = pd.concat((pd.read_csv(file, on_bad_lines='skip') for file in csv_files), ignore_index=True)
    return final_df
```

```python
# Example usage:
folder_path = '/content/sample_data/csv'
final_df  = concat_csv_files(folder_path)
```

```python
market_popularity = (final_df.album_available_markets.str.split(',')
                      .explode()
                      .value_counts()
                      .sort_values(ascending=False))
market_popularity.head(3)  # top 3 markets
```

```python
genre_popularity = (final_df.artist_genres.str.split(',')
                      .explode()
                      .value_counts()
                      .sort_values(ascending=False))
genre_popularity.head(3)    # top 3 genres
```
